```python
# coding=utf8
import xml.etree.ElementTree as etree
import time
import os
import re
import simplejson as json
from codecs import open

from settings import DATA_FOLDER
SOURCE_FILE = os.path.join(DATA_FOLDER, 'enwiki-20170820-pages-articles.xml')
OUTPUT_FILE = os.path.join(DATA_FOLDER, 'test_links.txt')
INDEX_FILE = os.path.join(DATA_FOLDER, 'test_index.json')
LINK_PATTERN = re.compile("\[\[(.+?)\]\]")
CUTOFF_PATTERN = re.compile("\s==[\sA-Za-z]")


def strip_tag_name(t):
    t = t.split('}')[1]
    return t


def article_generator(file):

    redirect = False
    not_article = False
    yielded = 0

    generator = etree.iterparse(file, events=('start', 'end'))
    _, root = generator.next()

    for event, elem in generator:
        xml_tag = strip_tag_name(elem.tag)

        if event == 'start':
            if xml_tag == 'page':
                title = None
                article_id = None
                text = None
                revision = False
                redirect = False
                not_article = False
            elif xml_tag == 'revision':
                revision = True

        else:
            if not_article:
                continue
            if xml_tag == 'title':
                title = elem.text
                if ":" in title or "(disambiguation)" in title:
                    not_article = True
            if xml_tag == 'id' and not revision:
                article_id = int(elem.text)
            if xml_tag == 'redirect':
                text = elem.attrib['title']
                redirect = True
            if not redirect and xml_tag == 'text':
                text = elem.text
                if not "===" in text:
                    not_article = True
            if xml_tag == 'ns' and elem.text != '0':
                not_article = True
            if xml_tag == 'page':
                if title and article_id and text:
                    yield title, article_id, text, redirect
                    yielded += 1
                    root.clear()


def clean_links(links):
    return [link.split('|')[0] for link in links if ":" not in link and link[0] != "#"]


def cutoff_position(text):
    match = re.search(CUTOFF_PATTERN, text)
    if match:
        return match.span()[0]
    else:
        return None
```

```python
start_time = time.time()
article_indexes = {}
redirect_references = {}
articles = 0
cutoff_articles = 0
with open(OUTPUT_FILE, 'w', encoding='utf-8') as f:
    for i, (title, id, text, redirect) in enumerate(article_generator(SOURCE_FILE)):
        if i % 10000 == 0 and i > 0:
            print str(articles) + " articles"
            print "ratio = " + str(cutoff_articles/float(articles))
            print time.time() - start_time
        if not redirect:
            articles += 1
            if re.search(CUTOFF_PATTERN, text):
                cutoff_articles += 1
            else:
                print text
                print i
                print title
                print
                print 
                print
start_time = time.time()
```
